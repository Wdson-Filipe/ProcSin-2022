```python

# Importar dados############################################################################################################

import pandas as pd, numpy as np, scipy as sp
#dadosLaboratorio = "TEK0013.CSV"
dadosLaboratorio = r'https://raw.githubusercontent.com/Wdson-Filipe/ProcSin-2022/master/TEK0013.CSV'
dados = pd.read_csv(dadosLaboratorio)
dados.head(10)
```

```python

#Função  FFT###############################################################################################################

import numpy as np,scipy as sp, matplotlib.pyplot as plt
from scipy.fft import fft #fast Fourier Transform
from scipy import signal

def fourier(x, t, Hz=False, dB=False, deg=False, Nfft=None):
    
    N = len(x); # Numero de amostras
    T = np.mean(t[1:]-t[:-1])# Periodo de amostragem 
     
    if Nfft == None:
        Nfft = N      
    
    #Estrutura do codigo detirada do exemplo dado em sala 
    #<https://github.com/ifmg-betim/2022.2_AUT.040_ProcSin/blob/main/aulas/ProcSin%20-%20002%20-%20FFT.ipynb>
    
    X_f = fft(x, Nfft)
    X_f = X_f/N #Corrige a amplitude do sinal
    X_f = X_f[:Nfft//2] #Suprime espelhamentos do sinal na frequencia 
    X_f[1:] = 2*X_f[1:];#Dobra a amplitude da frequencia diferente de zero

    
    if Hz == False:
        
        w0 = 2*np.pi/(Nfft*T); #Frequencia em rad/s
        
    else:
        
        w0 = 1/(Nfft*T); #Frequencia em Hz
    f = np.arange(Nfft)*w0; #Frequencia em rad/amostra
    f = f[0:Nfft//2];
    
    if dB == False: 
        
        mag = np.abs(X_f); 
        
    else: 
    
        mag = 20*np.log10(np.abs(X_f)); #Modulo no dominio da frequencia em dB
    pha = np.angle(X_f, deg=deg); #Fase
    
    return mag, pha, f;

### FIM DO CÓDIGO ###
```
